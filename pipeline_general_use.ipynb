```python
# packages
from transformers import pipeline

```

```python
# only provide task
pipe = pipeline(task="text-classification")
```

```python
#  run pipe
pipe("I like it very much.")
```

```python
# provide model
pipe = pipeline(task="text-classification",
                model="nlptown/bert-base-multilingual-uncased-sentiment")
```

```python
# run pipe
# consume just a string
pipe("I like it very much.")

```

```python
# consume a list
pipe(["I like it very much.",
      "I hate it."])


```
